### 1.- Class Imbalance:
Naturally there is not an equal number of healty and unhelthy data examples in medical datasets, this is highly corelated to the prevalence of an especific disease. There tend to be more examples of normal than anromal data, especially for X-ray studies. Sometimes this ratio can be 100/1 or even 1000/1.

![prevalence](utils\prevalence.PNG)

The latter creates a problem to the learning algorithm, since it would mostly see normal examples. This yields a model that starts ro predict very low probability of disease for everybody and won't be able to identify when an example has a disease. 

This can be traced back to the Binary Cross Entropy Loss Function (BCE)

$$ BCE = - \frac{1}{N} \sum_{i=1}^{N} y_i*log(P(y_i|x_i))\;+\;(1-y_i)*log(1-P(y_i|x_i)) $$

Since we are dealing with probabilites we have to take into account the three axioms of probability:

$AXIOM \; I) \;\;\;\;  P(A) \geq  0 \;\; \forall A \in S$

$AXIOM \; II)\;\; P(S) = S$

$AXIOM \; III)\; IF \;A \; \cap \;B \; then \; PI(A \cap B) =  P(A) + P(B) $

This means that the model prediction belongs to a tiny subset of the real numbers :$ P(y_i|x_i) \in R^{1}\in [0,1] $

If you are note too comfortable with conditional probability we could write $BCE$ in the follwoing way, just take into account that the prediction still needs to be in the subset we have specified previously:

$$ BCE = - \frac{1}{N} \sum_{i=1}^{N} y_i*log(\hat{y_i}(x_i))\;+\;(1-y_i)*log(1-\hat{y}(x_i)) $$

Where: $y_i$ is the actual label and $\hat{y_i}$ is the model's prediction that depends on $ x_i $ the vector or matrix of features. Notice that the BCE  exibits differnt behavior when $y_i=1$ and $y_i=0$

$BCE$ if $y_i=1$
$$ BCE = - \frac{1}{N} \sum_{i=1}^{N} 1*log(\hat{y_i}(x_i))\;+\;(1-1)*log(1-\hat{y}(x_i)) =  - \frac{1}{N} \sum_{i=1}^{N} 1*log(\hat{y_i}(x_i))\;+\; 0$$


$BCE$ if $y_i=0$
$$ BCE = - \frac{1}{N} \sum_{i=1}^{N} 0*log(\hat{y_i}(x_i))\;+\;(1-0)*log(1-\hat{y}(x_i)) =  - \frac{1}{N} \sum_{i=1}^{N} 0 \;+\;(1)*log(1-\hat{y}(x_i) $$

**Reamarks:**

Notice that in  $BCE$ with $y_i=1$ if $\hat{y}$ is labeled as 0 (missed labeled) the loss function tends to infinity. In the other hand, if $\hat{y}$  is labled as 1 (correctly labeled) the loss function tends to zero

Similarly in  $BCE$ with $y_i=0$ if $\hat{y}$ is labeled as 0 (correctly labeled) the loss function tends to zero. In the other hand, if $\hat{y}$  is labled as 1  (missed labeled) the loss function tends to infinity

Since Medical data can be in the ration of 100/1, it would be natural for the  binary crossentropy loss function to learn that all the inputs are healthy patients. Remeber That the loss over an epoch is the sumation over all examples's losses, meaning that if we have a missed mashed we would still get a fairly low loss.

### Solutions to the  Class Imbalance problem:

#### 1.1- Weighted Binary Cross Entropy:

$$ WBCE = - \frac{1}{N} \sum_{i=1}^{N} w_py_i*log(\hat{y_i}(x_i))\;+\;w_n(1-y_i)*log(1-\hat{y}(x_i)) $$

Where:
$$ w_p = \frac{N_{negative}}{N_{negative}+ N_{positive}} $$

$$ w_n = \frac{N_{positive}}{N_{negative}+ N_{positive}} $$

For instance if we have a dataset of 10 images with 9 normal and 1 abnromal patient:

$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$$ w_p = \frac{1}{10}=0.1 $ $\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ $ w_n = \frac{9}{10}=0.9 $

This will balance the loss so that the algorithm learns evenly.

#### EXAMPLE:

Suppose we have a dataset of 4  patients 3 healthy and 1 unhealthy and we have two models

- Model 1 always outputs a 0.9 for any example that it's given.  
- Model 2 always outputs a 0.1 for any example that it's given.

```python
import numpy as np
```

```python
def binary_crossentropy(y,y_hat):
    ''' Computes unweighetd binary crossentropy'''
    l = -(y.dot(np.log(y_hat.T)) +(1-y).dot(np.log(1-y_hat.T)))
    return l

def weighted_binary_crossentropy(y,y_hat,n=1,p=3):
    ''' Computes weighetd binary crossentropy'''
    wp = n/(n+p) # positive weight
    wn = p/(n+p) # Negative weight
    l = -( wp*y.dot(np.log(y_hat.T)) + wn*(1-y).dot(np.log(1-y_hat.T)))
    return l


# Generate an array of 4 binary label values, 3 positive and 1 negative
y_true = np.array([1,1,1,0])
print(f"y_true: \n{y_true}")

### Model 1 ###
y_pred_1 = 0.9 * np.ones(y_true.shape)
print(f"y_pred_1: \n{y_pred_1}")

### Model 2###
y_pred_2 = 0.1 * np.ones(y_true.shape)
print(f"y_pred_2: \n{y_pred_2}")
```

```python
### unweighted Binary Crossentropy ####
print("Unweighted Binary Crossentropy: ")
print(f"Loss of First model  with unweighted function: {binary_crossentropy(y_true,y_pred_1)}")
print(f"Loss of Second model with unweighted function: {binary_crossentropy(y_true,y_pred_2)}")
print()
### weighted Binary Crossentropy ####
print("Weighted Binary Crossentropy: ")
print(f"Loss of First  model with weighted function: {weighted_binary_crossentropy(y_true,y_pred_1,n=1,p=3)}")
print(f"Loss of Second model  with weighted function: {weighted_binary_crossentropy(y_true,y_pred_1,n=1,p=3)}")
```

**Remark:**
<br>
Notices that the unweighted loss tells us that the model that predicts every patient as healthy performs better than the one that tells us that the patient is unhealthy (due to class inbalance). However once that we set the weighted loss both models perform equally, meaning that we have succesfully achived our goal. an Important note is that weighted loss is only used during training, once our model is ready to deploy there is no need to weight the loss again (and it would not be practical since we could not predict how many patients of each class we are getting on our day to day life).

### 2.- Multi-Task

Basic image recongiation task focuse on just one task,  However, in Medical AI, we care about classifying the presence or absence of many  diseases. Now, one simple way to do this, is to have models that each learn one of these tasks. Or  maybe we can learn to do all of the tasks using one model. An advantage of this is that we can learn features that are common to identifying more than one disease, allowing us to use our existing data more efficiently. This is the setup of multitask learning. 

###  3.- Dataset Size
The challenge is that all Neural Netowrk Architectures are data hungry and benefit from the millions of examples found in image classification datasets. In medical problems, how can we still apply these techniques when we don't have millions of examples? One solution, is to pre-train the network. Here the idea is to first have the network, look at natural images, and learn to identify objects such as penguins, or cats, or dogs, then use this network as a starting point for learning in medical imaging task by copying over the learned features. The network can then further be trained to look at chest X-rays and identify the presence and absence of diseases. The idea of this process, is that when we're learning our first task of identifying cats or dogs, the network will learn general features that will help it's learning on the medical task. An example of this, might be that the features that are useful to identify the edges on a penguin, are also useful for identifying edges on a lung, which are then helpful to identify certain diseases. Then when we transfer these features to our new network, the network can learn the new task of chest X-ray interpretation with a better starting point. This first step, is called pre-training, and the second step, is called fine-tuning. It is generally understood that the early layers of the network, capture low-level image features that are broadly generalizable, while the later layers capture details that are more high-level or more specific to a task. 

As in any machine learning algorithm, it is a good practice to divide the dataset into  Training  and Testing set, but it is even better to subdivide the Training set into the Training and Validation set, it is important to shuffle the data so that all datasets have the same distribution:

![Datasets](utils/datasets.PNG)

* **Training set:** Used to train the algorithm on a per epoch bases
* **Validation set:** Used to _validatae_ the algorithm is learning after every epoch, this helps us apply earlystopping (if we are performing worse)
* **Testing set:** Used to evaluate how well the algorithm performs after trainning

It also important  to keep in mind that in a medical Dataset there may multiple exams of the same patient, this can be repeated infromation, it is recomend  that a patient is only found in one Dataset not in the 3 datasetsm this to verify that the information is unique (by Patient ID).

**Example:**

```python
import pandas as pd
# Read csv file containing training data
train_df = pd.read_csv("nih/train-small.csv")
# Print first 5 rows
print(f'There are {train_df.shape[0]} rows and {train_df.shape[1]} columns in the training dataframe')
train_df.head()
```

```python
# Read csv file containing validation data
valid_df = pd.read_csv("nih/valid-small.csv")
# Print first 5 rows
print(f'There are {valid_df.shape[0]} rows and {valid_df.shape[1]} columns in the validation dataframe')
valid_df.head()
```

```python
# Extract patient id's for the training set
ids_train = train_df.PatientId.values
# Extract patient id's for the validation set
ids_valid = valid_df.PatientId.values

# Create a "set" datastructure of the training set id's to identify unique id's
ids_train_set = set(ids_train)
print(f'There are {len(ids_train_set)} unique Patient IDs in the training set out of the {train_df.shape[0]} X-rays')
# Create a "set" datastructure of the validation set id's to identify unique id's
ids_valid_set = set(ids_valid)
print(f'There are {len(ids_valid_set)} unique Patient IDs in the validation set out of the {valid_df.shape[0]} X-rays')
```

```python
# Identify patient overlap by looking at the intersection between the sets
patient_overlap = list(ids_train_set.intersection(ids_valid_set))
n_overlap = len(patient_overlap)
print(f"There are {n_overlap} overlaps in trainning and validation sets. Id's of overlaping patients: \n{patient_overlap}")
```

```python
train_overlap_idxs = []
valid_overlap_idxs = []
for idx in range(n_overlap):
    train_overlap_idxs.extend(train_df.index[train_df['PatientId'] == patient_overlap[idx]].tolist())
    valid_overlap_idxs.extend(valid_df.index[valid_df['PatientId'] == patient_overlap[idx]].tolist())
    
print(f'These are the indices of overlapping patients in the training set: ')
print(f'{train_overlap_idxs}')
print(f'These are the indices of overlapping patients in the validation set: ')
print(f'{valid_overlap_idxs}')
```

```python
# Drop the overlapping rows from the validation set
valid_df.drop(valid_overlap_idxs, inplace=True)

# Extract patient id's for the validation set
ids_valid = valid_df.PatientId.values
# Create a "set" datastructure of the validation set id's to identify unique id's
ids_valid_set = set(ids_valid)
print(f'There are {len(ids_valid_set)} unique Patient IDs in the validation set')

# Identify patient overlap by looking at the intersection between the sets
patient_overlap = list(ids_train_set.intersection(ids_valid_set))
n_overlap = len(patient_overlap)
print(f'There are {n_overlap} Patient IDs in both the training and validation sets')
```

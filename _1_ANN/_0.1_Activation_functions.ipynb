```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
#Activation Functions:
def Sigmoid(Z):
    s = 1/(1+np.exp(-Z))
    return s

def Tanh(Z):
    t = np.tanh(Z)
    return t

def ReLU(Z):
    return Z * (Z > 0)

def Leaky_ReLU(Z):
    return Z * (Z > 0) + 0.1*Z * (Z < 0)
```

```python
# Derivatives of Activation Functions:

def dSigmoid(Z):
    return Sigmoid(Z)*(1-Sigmoid(Z))

def dTanh(Z):
    return 1-Tanh(Z)**2

def dReLU(Z):
    return 1. * (Z > 0)

def dLeaky_ReLU(Z):
    return 1* (Z > 0) + 0.1*(Z < 0)
```

```python
Z  =np.linspace(-5,5,102)
```

### MOST COMMON ACTIVATION FUNCTIONS

```python
plt.rcParams['figure.figsize'] = [12,17]
fig = plt.figure()
##### SIGMOID #####
ax1 = fig.add_subplot(421)
ax1.plot(Z,Sigmoid(Z))
ax1.set_title("SIGMOID")
ax1.set_ylim(-1.1,1.1)
plt.grid()

ax5 = fig.add_subplot(422)
ax5.plot(Z,dSigmoid(Z),"g")
ax5.set_title("Derivative ofSIGMOID")
ax5.set_ylim(-1.1,1.1)
plt.grid()

##### TANH #####
ax2 = fig.add_subplot(423)
ax2.plot(Z,Tanh(Z))
ax2.set_title("TANH")
ax2.set_ylim(-1.1,1.1)
plt.grid()

ax6 = fig.add_subplot(424)
ax6.plot(Z,dTanh(Z),"g")
ax6.set_title("Derivate of TANH")
ax6.set_ylim(-1.1,1.1)
plt.grid()

#### Relu ####

ax3 = fig.add_subplot(425)
ax3.plot(Z,ReLU(Z))
ax3.set_title("ReLU")
plt.grid()

ax7 = fig.add_subplot(426)
ax7.plot(Z,dReLU(Z),"g")
ax7.set_title("Derivate of TANH")

plt.grid()


#### Leaky Relu ####
ax4 = fig.add_subplot(427)
ax4.plot(Z,Leaky_ReLU(Z))
ax4.set_title("Leaky ReLU")
plt.grid()

ax8 = fig.add_subplot(428)
ax8.plot(Z,dLeaky_ReLU(Z),"g")
ax8.set_title("Leaky ReLU")

plt.grid()


plt.show()
```

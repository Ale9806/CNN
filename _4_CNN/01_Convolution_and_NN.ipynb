# Convolutional Neural Networks

In deep learning, a convolutional neural network (CNN, or ConvNet) is a class of deep neural networks, most commonly applied to analyzing visual imagery. CNN's relay on the basic idea of fully connected or ANN's, in fact, the last layers of CNN's are fully connected layers. The main difference with ANN's is that CCN's have two extra: Pooling layers and Convolutional Layers, the latter applies image convolutions with kernels, these  kernels update every iteration so that the CNN 'learns' filters that enables the network to differentiate from different classes. Before we begin lets have a short summary of ANN's. In summary, there are  three typical layers a Convolutional Neural Network has are:
* 1. Convolutilonal layer
* 2. Pooling layer
* 3. Fully connected layer

## Summary on  Artificial Neural Networks
Remeber that conventional neural networks are nothing more then a collection of  fundamental mathematical operations we can divide an  Artificial Neural Networks in two sections:
* Feedforward, this section is used in production and in the next step.
* Backforward, this section is only used when trainning a ANN

Given a dataset $ S =[(X_i,y_i),i=1,...,n]$, we define the Feedforward step of a  Neural Network as:


$ \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;
Z1 =  W_1A_0 + b_1 $
<br>
$ \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; A1 =  g(Z_1) $
<br>
$ \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; Z2 =  W_2A_1 + b_2 $
<br>
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; A2 =  g(Z_2) $
<br>
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; .$
<br>
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; .$
<br>
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; .$
<br>
$ \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; Zn =  W_nA_{n-1} + b_2 $
<br>
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\; An =  g(Zn) $

Where: 
$ A_0 = X  \;\;, \;\;   b_i \in R^{1} \;\;, \;\;  W_i \in R^{n,n-1}  \;\; and \;\; g() $ is a non-linear activation function

After this feedforward step, we train the Neural Network with a loss function that we define, It is important that this Loss function is monotonically decreasing so that for every iteration our system 'learns'. We find the best parameters $W^*$ for our NN minimizing the loss function, we do this using back-propagation and gradient descent.
<br>
$ \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;
W^* = argmin_w L(w) $

From our Calculus lectures we know that we can find the minimum or maximum point of a function with it's derivative. We also know that the a point is a maximum or a minimum if and only if:

$ \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;
\frac{\partial L}{\partial W} = 0 $

`Ideally` since we defined our loss function $L$ as a strictlly  monotonically decreasing function  we know that $ \frac{\partial L}{\partial W} = 0 $. In practice we will find a either a local or global mimimum.


**Example of backpropagation:** updating  $W$ and $b$ using Gradient descent and backpropagtion for the following Neural Network:
<font color=#000066> 
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;
W_1 \;\;\; b_1
\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;
W_2 \;\;\;\;\;\; b_2
$
<br>
$\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;
\downarrow \;\;\;\;\; \downarrow \;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;\;
\downarrow \;\;\;\;\;\;\;\; \downarrow$
\begin{equation}
X \to \boxed{Z_1 = W_1X +b_1} \to  \boxed{A_1 = g_1(Z_1)} \to \boxed{Z_2 = W_2A_1 +b_2} \to  \boxed{A_2 = g_2(Z_2)= \bar{y}} \to  \boxed{ L(y,\bar{y}=A_2)} \leftarrow y
\end{equation}
</font>
Learning rules:
$$ \frac{\partial L}{\partial W_2} = \frac{\partial L}{\partial A_2}*\frac{\partial A_2}{\partial Z_2}*\frac{\partial Z_2}{\partial W_2} $$

$$ \frac{\partial L}{\partial b_2} = \frac{\partial L}{\partial A_2}*\frac{\partial A_2}{\partial Z_2}*\frac{\partial Z_2}{\partial b_1} $$

$$ \frac{\partial L}{\partial W_1} = \frac{\partial L}{\partial A_2}*\frac{\partial A_2}{\partial Z_2}*\frac{\partial Z_2}{\partial A_1}*\frac{\partial A_1}{\partial Z_1}*\frac{\partial Z_1}{\partial W_1} $$

$$ \frac{\partial L}{\partial b_1} = \frac{\partial L}{\partial A_2}*\frac{\partial A_2}{\partial Z_2}*\frac{\partial Z_2}{\partial A_1}*\frac{\partial A_1}{\partial Z_1}*\frac{\partial Z_1}{\partial b_1} $$

Gradient Descent:
$$\bar{W_1} =  W_1 - \alpha*\frac{\partial L}{\partial W_1} $$
$$\bar{b_1} =  b_1 - \alpha*\frac{\partial L}{\partial b_1} $$
Gradient Descent:
$$\bar{W_2} =  W_2 - \alpha*\frac{\partial L}{\partial W_2} $$
$$\bar{b_2} =  b_2 - \alpha*\frac{\partial L}{\partial b_2} $$
Where  $ \alpha \in R^{1} $

We expect that in a finite number of iterations $\bar{W} \to W^*  $

**Pseudo code:**
``` python
for i in range(0,itreations):
    ### Feed Forward ###
    Z1 =  W1A0 + b_1
    A1 =  g(Z1)
    Z2 =  W2A1 + b_2 
    A2 =  g(Z2)
    
    ### Cost function ###
    L = l(y,A2)
    
    ### Back Propagation ###
    dL_dW2 = (dL_dA2)*(dA2_d_Z2)*(d_Z2_dW2)
    dL_db2 = (dL_dA2)*(dA2_d_Z2)*(d_Z2_db2)
    
    dL_dW1 = (dL_dA2)*(dA2_d_Z2)*(d_Z2_dA1)*(dA1_d_Z1)*(d_Z1_dW1)
    dL_db1 = (dL_dA2)*(dA2_d_Z2)*(d_Z2_dA1)*(dA1_d_Z1)*(d_Z1_db1)
    
    ### Update rule through Gradient Descent ###
    W2 = W2 - alpha*dL_dW2
    b2 = b2 - alpha*dL_db2
    W1 = W1 - alpha*dL_dW1
    b1 = b1 - alpha*dL_db1
   
```

## Back to CNN's:
With this in mind lets define the convolutional layer, for this is important to review the concpet of Convolution, in singals and systems we define them as: 

<font color=#006600> Continuos Time Convolution: $ f(f)*g(t) = \int_{-\infty}^{+\infty} f(\tau)g(t-\tau)d\tau  $ $\;\;\;\;\;\;\;\;\;\;\;\;\;\;$ </font>
<font color=#000099>Discrete Time Convolution:  $f[n]*g[n]  = \sum_{-\infty}^{+\infty} x[k]h[n-k]$ </font>

Convolutions for images are quite similar, and of course they are discrete since our `signal` is a collection of pixels stored in a tensor. We can have this tensor defined as $ T_{bw} \in R^{nxmx1}$ a (Black and white  photo) or as a tensor $T_{rgb} \in R^{nxmx3}$ (RGB photo)

### Valid Convolution:

Lets forget about $T_{rgb} $ and define convolution on an image for a  tensor $ T_{bw} \in R^{nxmx1}$ that we can define as a matrix $ M_{bw} \in R^{nxm}$:

In words, an image convolution for $ M_{bw} \in R^{nxm}$ will generate a matrix defined as $ M_{c} \in R^{(n-k)x(m-k)}$ the simplest form of image convolution (without padding) will return a matrix smaller than the orginal one,this type of convoultion is called `valid convolution`.

![convolution](utils\convolution.png)

Therefore we define the image  valid convolution as:
$$M_c[i_c,j_c] = M_{bw}[i,j]*F[f,f]  = \sum_{i=0}^{i_c}\sum_{j=0}^{j_c} M_{bw}[i:i+f,j:j+f] \; \otimes\;F[f,f]$$

Where $ \otimes $ Represents the element wise matrix mutliplication and:
$$ i_c = round({n-f}+1) $$
$$ j_c = round({m-f}+1) $$

With $n$: Matrix's height, $m$: Matrix's width, and $f$: kernels's length

Filters tend to be Squared and usally with odd-number dimensions 

Example of Valid Convolution:

````python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import matplotlib.cbook as cbook
import math

# colors for print()
class color:
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   BOLD = '\033[1m'
   END = '\033[0m'

print(color.GREEN + "9/9/2020"+ color.END)
````

````python
def show_mat(A,axis_n,ax):
    """Plots Matrix with numerical value"""
    
    ai = A.shape[0]  # Get Matrix Higth                                                
    aj = A.shape[1]  # Get Matrix Width
    norm = colors.SymLogNorm(linthresh=0.03,vmin=-20, vmax=20,base=10) # Normalize NUmerical values so that colors are consitant
    ax[axis_n].matshow(A, norm=norm,cmap=plt.cm.Blues)                 # Plot Matrix
    ax[axis_n].axis("off")                                             # Turn off axis
    ### Iiterate over the 
    for i in range(0,ai):     # Iterate over rows 
        for j in range(0,aj): # Iterate over Columns 
            c = A[i,j]        # Get current value
            ### Print value ###
            ax[axis_n].text(j, i, str(c), va='center', ha='center') 
    

def plot_matrix(matrix,kernel,Conv,conv_type="valid"):
    """ Plot Matrix, kernel and Convolution product"""
    
    ### Generate sublots ###
    fig, ax = plt.subplots(nrows=1, ncols=3,figsize=(12,7),gridspec_kw={'width_ratios': [3, 1,2]})
    
    ### Plot original matrix ###
    ax[0].set_title("Original")
    show_mat(matrix,0,ax)
    
    ### Plot kernel###
    ax[1].set_title("Filter")
    show_mat(kernel,1,ax)
    
    ### Plot Convolution ###
    ax[2].set_title(conv_type + " Convolution")
    show_mat(Conv,2,ax)
    
    plt.show()
    
def valid_convolion(matrix,kernel,prt=True):
    """ Compute Valide Convolutionl"""
    
    ### DEFINE PARAMETERS ###
    n=matrix.shape[0] # Dimension of matrix's hight 
    m=matrix.shape[1] # Dimension of matrix's width 
    f=kernel.shape[0] # Dimensions of filters (Filters are always symetricalby convention)
    print(color.BOLD +"Shape of Matrix:"+ color.END,f"{n}x{m}",color.BOLD + "Shape of filter:"+ color.END, f"{f}x{f}")
    
    ### Calcuate step size ###
    ic = int(math.floor(n - f +1))
    jc = int(math.floor(m - f +1))
    ### Creates the Convolution object with zeros ###
    C=np.ones((ic ,jc )) 
    print(color.BOLD +"Convolution expceted output size:"+ color.END,f" {ic} x {jc}")
    
    ### Apply Convolution ###
    for i in range(ic):     
        for j in range(jc): 
            C[i,j]=(matrix[i:i+f,j:j+f]*kernel).sum()
     
    ### Print Answer if user wants it ###
    if prt == True:
        print(C)
    return C
````

````python
a=[10,10,10,0,0,0]
b=[0,0,0,10,10,10]
M_bw   = np.array([a,a,a,b,b,b])                # Initiate Matrix
kernel = np.array([[1,1,1],[0,0,0],[-1,-1,-1]]) # Initiate Filter
Conv = valid_convolion(M_bw,kernel,prt=False)   # Do Valid Convolution
plot_matrix(M_bw,kernel,Conv)                   # Visualize Output
````

### Same Convolution

Since we will apply lots of convolutions, a probelm we can face is that at a finte number of layers our image can turned int a really small matrix, one to which we can no longer  apply the convolution operation, in order to solve this problem we can add a padding to our image.

Padding is bassically adding a number to the edges of the Matrix (Usually 0) in order to incrase its shape, so the output of the convolution operation is the orginal size of the matrix or in other words `the output of the convolution has the same size as the matrix` hence the name `same convolution`.


![Padding](utils\padding.png)

The convolution operation is the same, the only difference is the formula for  $i_c$ and $j_c$:

$$M_c[i_c,j_c] = M_{bw}[i,j]*F[f,f]  = \sum_{i=0}^{i_c}\sum_{j=0}^{j_c} M_{bw}[i:i+f,j:j+f] \; \otimes\;F[f,f]$$

Where $ \otimes $ Represents the element wise matrix mutliplication and:
$$ i_c = round({n+2p-f}+1) $$
$$ j_c = round({m+2p-f}+1) $$

With $n$: Matrix's height, $m$: Matrix's width, $f$: kernels's length, and $p$: Padding

````python
def padded_matrix(matrix,pad=(2,2),prt=True):
    ''' Pads a Matrix'''
    space = "  "  # Space for format strings

    shape = matrix.shape # Get Matrix's shape
    n  = shape[0]        # Store Height
    m  = shape[1]        # Store Width
    n_p = pad[0]         # Store Height's padding
    m_p = pad[1]         # Store Width's padding

    ### Create a Matrix of zeros for padding ###
    padded_matrix = np.zeros((n+2*n_p,m+2*m_p))

    ### Interact with user ###
    padded_shape =  padded_matrix.shape 
    if prt == True:
        print(f" Original Matrix's shape: {n}x{m}")
        print(f" Padded {space }Matrix's shape: {padded_shape[0]}x{padded_shape[1]} with padd : {n_p}x{m_p}")

    ### Save the value of matrix ###
    for i in range(n_p,n+n_p):                     # iterate over height (centered at pad)
        for j in range(m_p,m+m_p):                 # iterate over width  (centered at pad)
            padded_matrix[i,j]=matrix[i-n_p,j-m_p] # save cented values on padded matrix

    return padded_matrix


def same_convolution(matrix,kernel,prt=True):
    """ Compute Valide Convolutionl"""
    
    ### DEFINE PARAMETERS ###
    n=matrix.shape[0] # Dimension of matrix's hight 
    m=matrix.shape[1] # Dimension of matrix's width 
    f=kernel.shape[0] # Dimensions of filters (Filters are always symetricalby convention)
    print(color.BOLD +"Shape of Matrix:"+ color.END,f"{n}x{m}",color.BOLD + "Shape of filter:"+ color.END, f"{f}x{f}")
    
    ### Calcuate step size and padd ###
    p1=0                                 # Inititate padding for height with 0
    p2=0                                 # Inititate padding for width  with 0
    ic = int(math.floor(n +2*p1 - f +1)) # Calculate  expected output height size
    jc = int(math.floor(m +2*p2 - f +1)) # Calculate  expected output width size
    while ic != n and  jc !=  m:         # While expected output and input do  not share same size
        p1+=1                            # Increment padding
        p2+=1                            # Increment padding
        ic = int(math.floor(n +2*p1 - f +1)) # Calculate  expected output height size
        jc = int(math.floor(m +2*p2 - f +1)) # Calculate  expected output width size
        
    ### Padd Matrix ###
    matrix_padded = padded_matrix(matrix,pad=(p1,p2))
        
    ### Creates the Convolution object with zeros ###
    C=np.ones((ic ,jc )) 
    print(color.BOLD +"Convolution expceted output size:"+ color.END,f" {ic} x {jc}")
    
    
    ### Apply Convolution ###
    for i in range(ic):     
        for j in range(jc): 
            C[i,j]=(matrix_padded[i:i+f,j:j+f]*kernel).sum()
     
    ### Print Answer if user wants it ###
    if prt == True:
        print(C)
    return C



def plot_matrix2(matrix,padded):
    """ Plot Matrix, kernel and Convolution product"""
    
    ### Generate sublots ###
    fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(12,7),gridspec_kw={'width_ratios': [2,3.4]})
    
    ### Plot original matrix ###
    ax[0].set_title("Original")
    show_mat(matrix,0,ax)
    
    ### Plot kernel###
    ax[1].set_title("Padded")
    show_mat(padded,1,ax)

    
    plt.show()
    
````

````python
a=[10,10,10,1,1,1]
b=[1,1,1,10,10,10]
M_bw   = np.array([a,a,a,b,b,b])                # Initiate Matrix
M_bw_padded = padded_matrix(M_bw,pad=(1,2))     # Padd Matrix
plot_matrix2(M_bw ,M_bw_padded)                 # View original and padded matrix 
````

````python
a=[10,10,10,0,0,0]
b=[0,0,0,10,10,10]
M_bw   = np.array([a,a,a,b,b,b])                # Initiate Matrix
kernel = np.array([[1,1,1],[0,0,0],[-1,-1,-1]]) # Initiate Filter
Conv = same_convolution(M_bw,kernel,prt=False)  # Do Same Convolution
plot_matrix(M_bw_padded,kernel,Conv,conv_type="same")  # Visualize Output
````

### Stride of a Convolution:
We define the stride of a convolution as the step size for every coeffciant operation, as before the convolution operation does not change, only the formula for  $i_c$ and $j_c$:

$$M_c[i_c,j_c] = M_{bw}[i,j]*F[f,f]  = \sum_{i=0}^{i_c}\sum_{j=0}^{j_c} M_{bw}[i:i+f,j:j+f] \; \otimes\;F[f,f]$$

Where $ \otimes $ Represents the element wise matrix mutliplication and:
$$ i_c = round(\frac{n+2p_1-f}{s}+1) $$

$$ j_c = round(\frac{m+2p_2-f}{s}+1) $$

With $n$: Matrix's height, $m$: Matrix's width, $f$: kernels's length, $p$: Padding, and $s$: Stide 

````python
def stride_convolution(matrix,kernel,s=2,prt=True,conv_type="valid"):
    """ Compute stride convolution with either valid or same  Convolutionl"""
    
    ### DEFINE PARAMETERS ###
    n=matrix.shape[0] # Dimension of matrix's hight 
    m=matrix.shape[1] # Dimension of matrix's width 
    f=kernel.shape[0] # Dimensions of filters (Filters are always symetricalby convention)
    print(color.BOLD +"Shape of Matrix:"+ color.END,f"{n}x{m}",color.BOLD + "Shape of filter:"+ color.END, f"{f}x{f}")
    print(color.BOLD +"Stride:"+ color.END,f"{s}")
    
    ### Calcuate step size and padd ###
    p1=0                                 # Inititate padding for height with 0
    p2=0                                 # Inititate padding for width  with 0
    ic = int(math.floor((n +2*p1 - f)/s +1)) # Calculate  expected output height size
    jc = int(math.floor((m +2*p2 - f)/s +1)) # Calculate  expected output width size
    
    ##### SAME CONVOLUTION #####
    if conv_type == "same":
        while ic != n:                               # While expected output and input do  not share same height
            p1+=1                                    # Increment padding
            ic = int(math.floor((n +2*p1 - f)/s +1)) # Calculate  expected output height size
            
        while jc !=  m:                              # While expected output and input do  not share same width
            p2+=1                                    # Increment padding
            jc = int(math.floor((m +2*p2 - f)/s +1)) # Calculate  expected output width size
            

        ### Padd Matrix ###
        matrix_padded = padded_matrix(matrix,pad=(p1,p2))
        
    ##### VALID CONVOLUTION #####
    elif conv_type == "valid":
        matrix_padded = matrix
        
    ### Creates the Convolution object with zeros ###
    C=np.ones((ic ,jc )) 
    print(color.BOLD +"Convolution expceted output size:"+ color.END,f" {ic} x {jc}")
    
    
    ### Apply Convolution ###
    for i in range(ic):     
        for j in range(jc): 
            C[i,j]=(matrix_padded[i*s:i*s+f,j*s:j*s+f]*kernel).sum()
     
    ### Print Answer if user wants it ###
    if prt == True:
        print(C)
    return C
````

### Valid Convolution
Input size == outut size

This does not mean that all the layers of the kernel have to be the same type of filter, the volume cna have different filters.

````python
def volume_convolution(image,kernel,s=2,prt=True,conv_type="valid"):
    
    ### Define parameters ####
    n=image.shape[0] # Dimension of image's hight 
    m=image.shape[1] # Dimension of image's width 
    c=image.shape[2] # Dimension of image's chanel
    
    f=kernel.shape[0]  # Dimensions of filters (Filters are always symetricalby convention)
    fc=kernel.shape[2] # Dimensions of filter's channel 
    
    print(color.BOLD +"Shape of Image:"+ color.END,f"{n}x{m}x{c}",color.BOLD + "Shape of filter:"+ color.END, f"{f}x{f}x{fc}")
    print(color.BOLD +"Stride:"+ color.END,f"{s}")
    
    r = img_array[:,:,0]
    g = img_array[:,:,1]
    b = img_array[:,:,2]
    
    ### Calcuate step size and padd ###
    p1=0                                 # Inititate padding for height with 0
    p2=0                                 # Inititate padding for width  with 0
    ic = int(math.floor((n +2*p1 - f)/s +1)) # Calculate  expected output height size
    jc = int(math.floor((m +2*p2 - f)/s +1)) # Calculate  expected output width size
    
    ##### SAME CONVOLUTION #####
    if conv_type == "same":
        while ic != n:                               # While expected output and input do  not share same height
            p1+=1                                    # Increment padding
            ic = int(math.floor((n +2*p1 - f)/s +1)) # Calculate  expected output height size
            
        while jc !=  m:                              # While expected output and input do  not share same width
            p2+=1                                    # Increment padding
            jc = int(math.floor((m +2*p2 - f)/s +1)) # Calculate  expected output width size
            

        ### Padd r,g,b ###
        r_padded = padded_matrix(r,pad=(p1,p2))
        g_padded = padded_matrix(g,pad=(p1,p2))
        b_padded = padded_matrix(b,pad=(p1,p2))
        
    #### VALID CONVOLUTION #####
    elif conv_type == "valid":
        r_padded = r
        g_padded = g
        b_padded = b
        
    ### Creates the Convolution object with zeros ###
    C=np.ones((ic ,jc,3 )) 
    print(color.BOLD +"Convolution expceted output size:"+ color.END,f" {C.shape}")
    
    ### Convolution ####
    ### Apply Convolution ###
    for i in range(ic):     
        for j in range(jc): 
            C[i,j,0]=int((r_padded[i*s:i*s+f,j*s:j*s+f]*kernel[:,:,0]).sum()) # Convolution on red   channel
            C[i,j,1]=int((g_padded[i*s:i*s+f,j*s:j*s+f]*kernel[:,:,1]).sum()) # Convolution on green channel
            C[i,j,2]=int((b_padded[i*s:i*s+f,j*s:j*s+f]*kernel[:,:,2]).sum()) # Convolution on blue  channel
      
    ### Print Answer if user wants it ###
    if prt == True:
        print(C)
    return C

def plot_images(img_array,Convoluted_image):
    
    ### Generate sublots ###
    fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(30,20))
    
    ### Plot original matrix ###
    ax[0].set_title("Original", fontsize=18)
    ax[0].imshow(img_array)
    ax[0].axis("off")
    
    #### Plot convoluted image ####
    ax[1].set_title("Convoluted image", fontsize=18)
    ax[1].imshow(Convoluted_image)
    ax[1].axis("off")
    
    plt.show()
    
    fig, ax = plt.subplots(nrows=1, ncols=3,figsize=(30,20))
    #### Plot red channel ####
    ax[0].set_title("Red Channel with Horizontal edge filter", fontsize=18)
    ax[0].imshow(Convoluted_image[:,:,0])
    ax[0].axis("off")
    
    #### Plot green channel ####
    ax[1].set_title("Green Channel with Vertical edge filter", fontsize=18)
    ax[1].imshow(Convoluted_image[:,:,1])
    ax[1].axis("off")
    
    #### Plot blue channel ####
    ax[2].set_title("Blue Channel with average filter", fontsize=18)
    ax[2].imshow(Convoluted_image[:,:,2])
    ax[2].axis("off")
    
    plt.show()
````

## Pooling layer
CNN's have another type of layers called pooling layers, this layers are similar to convolutions, with the only excpetion that instead of multiplying by a kernel we get the max argument in a matrix slice. The pooling layer does not have parameters to learn. If you want you could add padding to maxpooling but is not a commun thing to do

$$M_c[i_c,j_c] = Max_{pooling}(M_{bw}[i,j]) = arg_{max}( M_{bw}[i:i+f,j:j+f]) $$

$$ i_c = round(\frac{n-f}{s}+1) $$

$$ j_c = round(\frac{m-f}{s}+1) $$

With $n$: Matrix's height, $m$: Matrix's width, $f$: kernels's length, and $s$: Stide 

````python
def max_pooling(matrix,s=2,prt=True):
    """ Compute stride convolution with either valid or same  Convolutionl"""
    
    ### DEFINE PARAMETERS ###
    n=matrix.shape[0] # Dimension of matrix's hight 
    m=matrix.shape[1] # Dimension of matrix's width 
    f=kernel.shape[0] # Dimensions of filters (Filters are always symetricalby convention)
    print(color.BOLD +"Shape of Matrix:"+ color.END,f"{n}x{m}",color.BOLD + "Shape of filter:"+ color.END, f"{f}x{f}")
    print(color.BOLD +"Stride:"+ color.END,f"{s}")
    
    ### Calcuate step size ###

    ic = int(math.floor((n - f)/s +1)) # Calculate  expected output height size
    jc = int(math.floor((m - f)/s +1)) # Calculate  expected output width size
    
        
   
    ### Creates the Convolution object with zeros ###
    C=np.ones((ic ,jc )) 
    print(color.BOLD +"Convolution expceted output size:"+ color.END,f" {ic} x {jc}")
    
    
    ### Apply Convolution ###
    for i in range(ic):     
        for j in range(jc): 
            C[i,j]=np.max(matrix[i*s:i*s+f,j*s:j*s+f])
     
    ### Print Answer if user wants it ###
    if prt == True:
        print(C)
    return C

def plot_matrix3(matrix,pooling):
    """ Plot Matrix, kernel and Convolution product"""
    
    ### Generate sublots ###
    fig, ax = plt.subplots(nrows=1, ncols=2,figsize=(12,7),gridspec_kw={'width_ratios': [2,3.4]})
    
    ### Plot original matrix ###
    ax[0].set_title("Original")
    show_mat(matrix,0,ax)
    
    ### Plot kernel###
    ax[1].set_title("Max Pooling")
    show_mat(pooling,1,ax)

    plt.show()
    
````

````python
### Max Pooling ###
a=[10,20,10,0,0,0]
b=[0,0,0,10,10,10]
M_bw    = np.array([a,a,a,b,b,b])        # Initiate Matrix
Pooling = max_pooling(M_bw,s=2,prt=True) # Pooling layer
plot_matrix3(M_bw,Pooling)               # Show output
````
